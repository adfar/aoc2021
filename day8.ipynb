```python
signals = []
output = []
```

```python
with open("aoc2021/input8") as f:
    for line in f:
        line = line.strip().split('|')
        signals.append(line[0].strip().split(' '))
        output.append(line[1].strip().split(' '))
```

```python
maps = {
    0: None,
    1: None,
    2: None,
    3: None,
    4: None,
    5: None,
    6: None,
    7: None,
    8: None,
    9: None
}
```

```python
def decode(store, signals):
    store = store.copy()
    signals = signals.copy()
    signals.sort(key=len)
    for i in range(len(signals)):
        signals[i] = set(signals[i])
    store[8] = signals.pop(9)
    store[4] = signals.pop(2)
    store[7] = signals.pop(1)
    store[1] = signals.pop(0)
    for signal in signals:
        if len(signal) == 6 and not store[1].issubset(signal):
            store[6] = signal
            signals.remove(signal)
    for signal in signals:
        if len(signal) == 5 and store[1].issubset(signal):
            store[3] = signal
            signals.remove(signal)
    for signal in signals:
        if len(signal) == 6 and store[3].issubset(signal):
            store[9] = signal
            signals.remove(signal)
    for signal in signals:
        if len(signal) == 6:
            store[0] = signal
            signals.remove(signal)
    for signal in signals:
        if len(signal) == 5 and signal.issubset(store[6]):
            store[5] = signal
            signals.remove(signal)
    store[2] = signals[0]
        
        
    return store
```

```python
def out(store, digits):
    store = store.copy()
    digits = digits.copy()
    reading = ''
    for i in range(len(digits)):
        digits[i] = set(digits[i])
    for digit in digits:
        for key, value in store.items():
            if digit == value:
                reading += str(key)
    return int(reading)
```

```python
total = 0
for i in range(len(signals)):
    code = decode(maps, signals[i])
    total += out(code, output[i])
print(total)
```
